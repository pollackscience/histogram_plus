# Simple Example

```python
import numpy as np
from histogram_plus import hist

%matplotlib inline
plt.rcParams['figure.figsize'] = (12,8)

np.random.seed(1001)
data = np.random.normal(0, 1, size=1000)

_ = hist(data, errorbars=True)
plt.title('Automatic Error Bar Generation')

plt.figure()
_ = hist(data, errorbars=True, histtype='marker')
plt.title('Marker Draw-style')
```
