```python
import numpy as np
import matplotlib.pyplot as plt
from histogram_plus.hist_funcs import hist

%matplotlib inline
plt.rcParams['figure.figsize'] = (12,8)
```

```python
np.random.seed(8008)
data = np.random.normal(0, 1, size=1000)
weights = np.random.uniform(2,4, size=1000)
weights2 = np.random.uniform(0.1,200, size=1000)
weights3 = np.random.uniform(0.1,0.3, size=1000)
data2 = np.random.normal(3, 1, size=1000)
data3 = np.random.normal(6, 1, size=1000)
data4 = np.random.normal(6, 1, size=100)
weights4 = np.random.uniform(1,2, size=100)
data_uni = np.random.uniform(1,10, size=1000)

data_mixed_uni = np.concatenate([np.random.uniform(1,10, size=5000), np.random.uniform(2,5, size=600),
                                 np.random.uniform(6,7, size=300)])
```

```python
types = ['bar', 'stepfilled', 'step', 'marker']
#types=['marker']
#norms = [False, True]
norms = [False]
#scale = [None, 0.5, 2, 'binwidth']
#scale = [None, 2, 'binwidth']
scale = [None, 'binwidth']
#bins = [30, 'auto', 100, 'blocks']
bins = [30, 'blocks']
ws = [weights]
for t in types:
    for n in norms:
        for s in scale:
            for b in bins:
                for iw, w in enumerate(ws):
                    plt.figure()
                    hist(data, bins=b, histtype=t, normed=n, scale=s, errorbars=True, alpha=0.5, label='d1')
                    plt.title('type: {0}, norm: {1}, scale: {2}, binning: {3}, weights: {4}'.format(t, n, s, b, iw))
                    plt.legend()
```
