```python
import numpy as np
import matplotlib.pyplot as plt
from histogram_plus.hist_funcs import hist

%matplotlib inline
plt.rcParams['figure.figsize'] = (12,8)
```

```python
data = np.random.normal(0, 1, size=1000)
data2 = np.random.normal(1, 1, size=1000)
```

```python
types = ['bar', 'stepfilled', 'step', 'marker']
norms = [False, True]
scale = [None, 0.5, 2, 'binwidth']
bins = [30, 'auto', 100]
for t in types:
    for n in norms:
        for s in scale:
            for b in bins:
                plt.figure()
                hist(data, bins=b, histtype=t, normed=n, scale=s, errorbars=True, alpha=0.5)
                plt.title('type: {0}, norm: {1}, scale: {2}, binning: {3}'.format(t, n, s, b))
```
