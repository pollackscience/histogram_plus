```python
import numpy as np
import matplotlib.pyplot as plt
from histogram_plus.hist_funcs import hist

%matplotlib inline
plt.rcParams['figure.figsize'] = (12,8)
```

```python
data = np.random.normal(0, 1, size=1000)
```

```python
types = ['bar', 'stepfilled', 'step', 'marker']
norms = [True, False]
scale = [None, 0.5, 2, 'binwidth']
for t in types:
    for n in norms:
        for s in scale:
            plt.figure()
            hist(data, bins=30, histtype=t, normed=n, scale=s, alpha=0.5)
            plt.title(' '.join([t,str(n),str(s)]))
```

```python
types = ['bar', 'stepfilled', 'step', 'marker']
norms = [False, True]
scale = [None, 0.5, 2, 'binwidth']
for t in types:
    for n in norms:
        for s in scale:
            plt.figure()
            hist(data, bins=30, histtype=t, normed=n, scale=s, alpha=0.5, errorbars=True)
            plt.title(' '.join([t,str(n),str(s)]))
```
