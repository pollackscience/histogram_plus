# Ratio Plots

Automatic generation of ratio plots.  More documentation coming soon.

```python
import numpy as np
from histogram_plus import ratio_plot
%matplotlib inline
plt.rcParams['figure.figsize'] = (8,6)

np.random.seed(1001)
data = np.random.normal(size=1000)
data2 = np.random.normal(size=1000)
```

```python
axes, hist1, hist2 = ratio_plot(
    dict(x=data, errorbars=True, histtype='marker'),
    dict(x=data2, errorbars=True),
    bins=20
)

axes[0].set_title('Simple Ratio Plot')
```
